```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
```

Importing  Data Files as Separate Dataframes

```python
df1 = pd.read_csv('1623535615-3006-IAJ9206.csv')
df2 = pd.read_csv('1624040490-2957-NAI1549.csv')
df3 = pd.read_csv('1624374799-3006-MYS2071.csv')
df4 = pd.read_csv('1624962426-3006-NAH4736.csv')
```

Analyzing basic info of all dataframes

```python
df1.info()
df2.info()
df3.info()
df4.info()
```

Calculating time and frequency columns in dataframe which will be used to plot graphs

```python
df1['time'] = np.arange(len(df1))*3006
df2['time'] = np.arange(len(df2))*2957
df3['time'] = np.arange(len(df3))*3006
df4['time'] = np.arange(len(df4))*3006
```

```python
df1['freq_microHz'] = 1/df1.time*1000000 
df2['freq_microHz'] = 1/df2.time*1000000 
df3['freq_microHz'] = 1/df3.time*1000000 
df4['freq_microHz'] = 1/df4.time*1000000 
```

## Plotting graph of IAJ9206 

```python
pyplot.plot(df1.time,np.sin(df1.x))
pyplot.plot(df1.time,np.sin((df1.y)))
pyplot.plot(df1.time,np.sin((df1.z)))
pyplot.title('Time Domain')
```

```python
pyplot.plot(df1.freq_microHz,np.sin((df1.x)))
pyplot.plot(df1.freq_microHz,np.sin((df1.y)))
pyplot.plot(df1.freq_microHz,np.sin((df1.z)))
pyplot.title('Freq Domain')
```

```python
df1.plot.scatter(x='freq_microHz',y='z')
```

```python
df1 = df1[df1["freq_microHz"]<1]
```

```python
pyplot.plot(df1.time,np.sin(df1.x))
pyplot.plot(df1.time,np.sin((df1.y)))
pyplot.plot(df1.time,np.sin((df1.z)))
pyplot.title('Time Domain')
```
